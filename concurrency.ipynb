```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
current_con = pd.read_csv('current_concurrency.csv')
```

```python
current_con
```

```python
noon_per_area = current_con.groupby('area_name')['noon_concurrency'].sum().reset_index()
```

```python
fig = px.bar(noon_per_area, x='area_id', y='noon_concurrency')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```
